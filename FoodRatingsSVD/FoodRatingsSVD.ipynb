## Singular Value Decomposition

Singular Value Decomposition is a generalisation of the Eigendecomposition method and is applicable on any type of Matrix.It's very intutive when dealing with finding hidden factors.

```python
import pandas as pd
import numpy as np
np.set_printoptions(precision=2)
```

```python
np.version.version
```

```python
dat = pd.read_csv('MyFoodRatings.csv')
dat = dat.drop(['Name'], axis=1)
dat
```

```python
U,s,VT = np.linalg.svd(dat, full_matrices = True)
```

```python
#Also known as the matrix that shows the effect of each row on the themes
U 
```

```python
U.shape
```

```python
b = ['A','B','C','D','E','F','G','H','I','J','K','L']
dt = pd.DataFrame(U,columns = b)
dt
```

```python
#The strength of the themes are represented in the diagonals of this matrix
s
```

```python
np.diag(s[:3])
```

```python
#The themes to columns matrix
VT
```

```python
b = ['Chicken','Mutton','Paneer','ChowMein','SpringRolls','Momo','Sushi','Ramen','Tempura']
dt = pd.DataFrame(VT,columns = b)
dt
```
